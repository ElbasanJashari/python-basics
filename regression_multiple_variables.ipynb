With vectorization

```python
import numpy as np
x = [1, 2, 3, 4, 5, 12, 35 ,4 ,7 ,58]
w = [3, 2, 7, 5 ,6, 7, 12, 5, 4, 2]
b = 2

f = np.dot(w,x) + b

print (f)
```

Without vectorization

```python
x = [1, 2, 3, 4, 5, 12, 35 ,4 ,7 ,58]
w = [3, 2, 7, 5 ,6, 7, 12, 5, 4, 2]
b = 2

f = 0

for j in range(0,10):
    f = f + w[j] * x[j]

f = f + b

print (f)
```
