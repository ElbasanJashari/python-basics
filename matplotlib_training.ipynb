```python
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 15, 30, 25]
y2 = [5, 15, 10, 20, 18]

# Add a label= to each line
plt.plot(x, y1, color="blue",  linestyle="-",  marker="o", label="Training Score")
plt.plot(x, y2, color="red",   linestyle="--", marker="s", label="Validation Score")

plt.title("Training vs Validation")
plt.xlabel("Epochs")
plt.ylabel("Score")

plt.legend()   # ← This draws the legend box on the chart

plt.show()
```

```python
np.random.seed(42)
data = np.random.randn(200)

# Try different bin counts side by side
fig, axes = plt.subplots(1, 3, figsize=(12, 4))  # 1 row, 3 charts

axes[0].hist(data, bins=5,  color="red",       edgecolor="black")
axes[0].set_title("5 bins (too few)")

axes[1].hist(data, bins=20, color="steelblue", edgecolor="black")
axes[1].set_title("20 bins (just right)")

axes[2].hist(data, bins=80, color="green",     edgecolor="black")
axes[2].set_title("80 bins (too many)")

plt.suptitle("Effect of Bin Count", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
np.random.seed(42)
data = np.random.randn(50)

data_1 = [1, 4, 8, 9, 10]
x1 = [1, 2, 3, 4, 5]
y1 = [10, 20, 15, 30, 25]

models = ["LVV", "PDK", "AAK", "LDK"]
accuracy = [50, 25, 17, 8]
colors = ["red", "blue", "steelblue", "green"]

fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].hist(data, bins=5,  color="red",       edgecolor="black")
axes[0].set_title("5 bins (too few)")

axes[1].plot(data_1, x1, color="blue",  linestyle="-",  marker="o", label="Training Score")
axes[1].plot(data_1, y1, color="red",   linestyle="--", marker="s", label="Validation Score")
axes[1].set_title("second one")

axes[2].bar(models, accuracy, color=colors)

plt.suptitle("Effect of Bin Count", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset directly from the internet
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)


plt.style.use("seaborn-v0_8")
plt.figure(figsize=(8, 5))

plt.hist(df["sepal_length"], bins=20, color="steelblue", edgecolor="black")

plt.title("Distribution of Sepal Length", fontsize=18)
plt.xlabel("Sepal Length (cm)", fontsize=13)
plt.ylabel("Count", fontsize=13)

plt.show()
```

```python
plt.style.use("seaborn-v0_8")
plt.figure(figsize=(8, 5))

# Split the dataframe into 3 species
setosa     = df[df["species"] == "setosa"]
versicolor = df[df["species"] == "versicolor"]
virginica  = df[df["species"] == "virginica"]

# Plot each species as its own scatter cluster
plt.scatter(setosa["sepal_length"],     setosa["sepal_width"],     color="red",    label="Setosa",     marker="o")
plt.scatter(versicolor["sepal_length"], versicolor["sepal_width"], color="blue",   label="Versicolor", marker="s")
plt.scatter(virginica["sepal_length"],  virginica["sepal_width"],  color="green",  label="Virginica",  marker="^")

plt.title("Sepal Length vs Width by Species", fontsize=18)
plt.xlabel("Sepal Length (cm)", fontsize=13)
plt.ylabel("Sepal Width (cm)", fontsize=13)
plt.legend(fontsize=12)

plt.show()
```
