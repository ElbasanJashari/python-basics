System of Linear Equations

```python
import numpy as np

A = np.array([[4, -3, 1],
        [2, 1, 3],
        [-1, 2, -5]], dtype=np.dtype(float))

B = np.array([-10, 0, 17], dtype=np.dtype(float))

x = np.linalg.solve(A, B)
print(x)
```

Calculating determinant

```python
d = np.linalg.det(A)
print(f"{d:.2f}")
```

What happens if the system has no unique solution?

```python
A_2= np.array([
        [1, 1, 1],
        [0, 1, -3],
        [2, 1, 5]
    ], dtype=np.dtype(float))

b_2 = np.array([2, 1, 0], dtype=np.dtype(float))

d_2 = np.linalg.det(A_2)
print(d_2)
```

```python
import numpy as np
A = np.array([
    [1, 1],
    [0, 1]
  ], dtype=np.dtype(float))

B = np.array([7, 4], dtype=np.dtype(float))

A_B = np.linalg.solve(A, B)
A_Det = np.linalg.det(A)
print(A_B)
print(A_Det)
```

```python
import numpy as np
A = np.array([
    [1, 1, 2],
    [0, 1, 3],
    [1, 0, 1]
  ], dtype=np.dtype(float))

B = np.array([7, 4, 9], dtype=np.dtype(float))

A_B = np.linalg.solve(A, B)
A_Det = np.linalg.det(A)
print(A_B)
print(A_Det)
```
