```python
import pandas as pd

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"

df = pd.read_csv(url)

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

Removing Missing Data

```python
df = df.dropna()
```

Fill missing ages with the average age.

```python
df["age"] = df["age"].fillna(df["age"].mean())
```

Finding Duplicate Data

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

Looking at Unique Values

```python
df["species"].unique()
```

```python
df["species"].value_counts()
```

Quick Visualization with Matplotlib

```python
import matplotlib.pyplot as plt

df["sepal_length"].hist()

plt.xlabel("In cm")
plt.ylabel("Nr of flowers")
plt.title("The spread of sepal length")
plt.show()
```

```python
plt.scatter(df["sepal_length"], df["sepal_width"])

plt.xlabel("sepal length")
plt.ylabel("sepal width")

plt.show()
```

```python
setosa = df[df["species"] == "setosa"]
versicolor = df[df["species"] == "versicolor"]
virginica = df[df["species"] == "virginica"]

plt.scatter(setosa["sepal_length"], setosa["sepal_width"], label="setosa")
plt.scatter(versicolor["sepal_length"], versicolor["sepal_width"], label="versicolor")
plt.scatter(virginica["sepal_length"], virginica["sepal_width"], label="virginica")

plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.legend()

plt.show()
```

Pandas has built-in plotting.

```python
df.plot(kind="scatter", x="sepal_length", y="sepal_width")

plt.show()
```

```python
df.groupby("species")["petal_length"].mean().plot(kind="bar", color=["blue", "green", "red"])

plt.ylabel("Average Petal Length")
plt.show()
```

```python
df.corr(numeric_only=True)
```

```python
plt.imshow(df.corr(numeric_only=True))
plt.colorbar()
plt.show()
```
